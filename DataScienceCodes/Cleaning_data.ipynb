```python
import pandas as pd
df = pd.read_csv('mine.csv')
for x in df.index:
    if df.loc[x, 'Duration'] > 120:
       df.loc[x, 'Duration'] = 45
print(df)
```

```python
#removing rows
import pandas as pd
df = pd.read_csv('mine.csv')
for x in df.index:
  if df.loc[x, "Duration"] > 120:
    df.drop(x, inplace = True)
print(df)
```
